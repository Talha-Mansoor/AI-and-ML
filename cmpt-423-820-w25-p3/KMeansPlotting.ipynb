```python
import json
import matplotlib.pyplot as plt
import os

# Ensure output directory exists
os.makedirs("doc/figure", exist_ok=True)

ks = [2, 5, 10, 20, 50]
distortions = []

for k in ks:
    with open(f"k{k}.json") as f:
        data = json.load(f)
        distortions.append(data["distortion"])

# Plot and save
plt.figure(figsize=(6, 4))
plt.plot(ks, distortions, marker='o')
plt.title("Distortion vs. K")
plt.xlabel("K (Number of clusters)")
plt.ylabel("Distortion")
plt.grid(True)
plt.tight_layout()
plt.savefig("doc/figure/distortion_vs_k.png", dpi=300)
plt.show()
```
