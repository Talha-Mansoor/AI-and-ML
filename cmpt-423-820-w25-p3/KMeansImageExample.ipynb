This notebooks visualizes the MNIST images as well as the centroids of the K-Means dataset.

```python
import plotly.express as px
import matplotlib.pyplot as plt

from mpl_toolkits.axes_grid1 import ImageGrid

from kmeans.kmeans import MNIST7K, KMeansResult


def plot_digits(data):
    num_plots = data.shape[0]
    fig = plt.figure(figsize=(num_plots, 10.*num_plots))
    grid = ImageGrid(fig, 111, nrows_ncols=(1, num_plots), axes_pad=0.1)
    for i in range(num_plots):
        grid[i].imshow(data[i].reshape((28,28)), cmap='gray', vmin=0, vmax=1)
        grid[i].set_xticks([])
        grid[i].set_yticks([])
    return fig
```

The following code plots the first six images of the MNIST dataset.

```python
dataset = MNIST7K.from_file()
fig = plot_digits(dataset.imgs[:6].copy())
```

```python
fig.savefig("doc/figure/mnist.pdf", bbox_inches='tight')
```

The following code plots the centroids from a K-Means clustering object.

```python
res = KMeansResult.from_file("k50.json")
fig = plot_digits(res.mu[:].copy())
```

```python
fig.savefig("doc/figure/centroids_6.pdf", bbox_inches='tight')
```
